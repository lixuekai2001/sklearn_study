```python
import pandas as pd
df=pd.read_csv('SMSSpamCollection',sep='\t',header=None)
print(df.head(5))
#read_csv万能读取
#df.head(N)读取前n行数据
```

```python
print('Number of spam messages: %s' %df[df[0]=='spam'][0].count())
print('Number of spam messages: %s' %df[df[0]=='ham'][0].count())
```

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model.logistic import LogisticRegression
from sklearn.model_selection import train_test_split,cross_val_score
```

```python
X=df[1].values
y=df[0].values
X_train_raw,X_test_raw,y_train,y_test=train_test_split(X,y)
vectorizer=TfidfVectorizer()
#可以把档转换成TF-IDF特征的矩阵
X_train=vectorizer.fit_transform(X_train_raw)
X_test=vectorizer.transform(X_test_raw)
print(X_train)
```

```python
for key,value in vectorizer.vocabulary_.items():
    print(key,value)
#每个出现的词的值
```

```python
classifier=LogisticRegression()
classifier.fit(X_train,y_train)
predictions=classifier.predict(X_test)
for i,prediction in enumerate(predictions):
    print('predicted: %s, target:%s' %(prediction,y_test[i]))
    print('predicted: %s, message:%s' %(prediction,X_test_raw[i]))
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
confusion_matrix=confusion_matrix(y_test,predictions)
print(confusion_matrix)
plt.matshow(confusion_matrix)
#plt.matshow是一个绘制矩阵的函数
plt.title('confusion matrix')
plt.colorbar()
plt.show()
```

```python
scores=cross_val_score(classifier,X_train,y_train,cv=5)
print('Acurracy :%s' %scores)
print('Mean accuracy: %s' %np.mean(scores))
```

```python
#y需要进行二值化
from sklearn.preprocessing import LabelBinarizer
Binarizer=LabelBinarizer()
y_train_binarizer=Binarizer.fit_transform(y_train)
y_test_binarizer=Binarizer.fit_transform(y_test)
print(y_train_binarizer)
```

```python
precisions=cross_val_score(classifier,X_train,y_train_binarizer.reshape(-1),cv=5,scoring='precision')
print('Mean precisions: %s' %np.mean(precisions))
recalls=cross_val_score(classifier,X_train,y_train_binarizer.reshape(-1),cv=5,scoring='recall')
print('Mean recall: %s' %np.mean(recalls))
f1s=cross_val_score(classifier,X_train,y_train_binarizer.reshape(-1),cv=5,scoring='f1')
print('Mean f1: %s' %np.mean(f1s))
# ?????????没找到哪有问题，可能输数据的问题
```

```python
from sklearn.metrics import roc_curve,auc
predictions=classifier.predict_proba(X_test)
#predict_proba返回的是一个n行k列的数组，第i行第j列上的数值是模型预测第i个预测样本为某个标签的概率，并且每一行的概率和为1
print(predictions)
```

```python
false_positive_rate,recall,thresholds=roc_curve(y_test_binarizer,predictions[:,1])
roc_auc=auc(false_positive_rate,recall)
plt.title('receiever oprating characteristic')
plt.plot(false_positive_rate,recall,'b',label='AUC=%0.2f' %roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([-0.01,1.0])
plt.ylim([0.0,1.0])
plt.xlabel('Fall-out')
plt.ylabel('Recall')
plt.show()
#ROC（Receiver Operating Characteristic）曲线和AUC常被用来评价一个二值分类器（binary classifier）的优劣。
#AUC（Area Under Curve）被定义为ROC曲线下的面积,越高值越好
#将截断点不同取值下对应的TPR和FPR结果画于二维坐标系中得到的曲线，就是ROC曲线
#一般来说，如果ROC是光滑的，那么基本可以判断没有太大的overfitting
#TPR=TP/(TP+FN)=TP/actual positives,就是recall
#FPR=FP/(FP+TN)=FP/actual negatives
#ROC是由点（TPR,FPR）组成的曲线，AUC就是ROC的面积。AUC越大越好
#截断点,对test样本进行预测后,可以输出各test样本对某个类别的相似度概率。比如t1是P类别的概率为0.3，一般我们认为概率低于截断点0.5，t1就属于类别N。
```
