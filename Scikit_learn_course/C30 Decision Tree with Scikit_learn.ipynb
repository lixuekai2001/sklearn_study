```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn import tree
#GridSearchCV被移到了model_selection模块中
```

```python
df=pd.read_csv('ad.data',header=None)
```

```python
print(df.head(200))
```

```python
explanatory_variable_columns=set(df.columns.values)
#set() 函数创建一个无序不重复元素集，可进行关系测试，删除重复数据，还可以计算交集、差集、并集等。
#因为源数据是没有特征标签的，因此利用set函数进行创建标签元素集
#固定用法，df.columns.values 返回列名
explanatory_variable_columns.remove(len(df.columns.values)-1)
#remove(1558)去出列前1558列，注意符合是小括号
response_variable_column=df[len(df.columns.values)-1]
#df[1558]采用索引办法因此是把最后一列作为response
```

```python
print(df.columns)
```

```python
print(explanatory_variable_columns)
```

```python
print(len(df.columns.values))
```

```python
print(response_variable_column)
```

```python
y=[1 if e=='ad.' else 0 for e in response_variable_column]
#非常好的写法，for 套在if 语句中
X=df[list(explanatory_variable_columns)].copy()
print(X)
```

```python
X.replace(to_replace=' *\?',value=-1,regex=True, inplace=True)
#空格*？：正则表达式，所谓-1取代为?,因为？正则里面属于特殊字符因此需要转义\?
#to_replace:正则表达式 regex=True 正则开启  inplace=True 替换开启
print(X)
X.describe()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
pipeline=Pipeline([
    ('clf',DecisionTreeClassifier(criterion='entropy'))
])
```

```python
parameters={
    'clf__max_depth':(150,155,160),
    'clf__min_samples_split':(2,3),
    'clf__min_samples_leaf':(1,2,3),     
}
```

```python
grid_search=GridSearchCV(pipeline,parameters,n_jobs=-1,verbose=1,scoring='f1')
grid_search.fit(X_train,y_train)
```

```python
best_parameters=grid_search.best_estimator_.get_params()
```

```python
print('Best score: %0.3f' %grid_search.best_score_)
print('Best parameter set:')
for param_name in sorted(parameters.keys()):
    print('\t%s:%r'%(param_name ,best_parameters[param_name])) 
predictions=grid_search.predict(X_test)
print(classification_report(y_test,predictions))
```

```python
clf=DecisionTreeClassifier(criterion='entropy',max_depth=155,min_samples_leaf=1,min_samples_split=3)
clf=clf.fit(X_train,y_train)
tree.export_graphviz(clf,out_file='tree.dot')
#绘制决策树图，需要graphviz
```
